```python
import networkx as nx
import matplotlib.pyplot as plt

# Nombres simulados de documentos
documentos = ["Doc1", "Doc2", "Doc3", "Doc4", "Doc5"]

# Simulaciones de similitud entre documentos
simulaciones = [
    ("Doc1", "Doc2", 78),
    ("Doc1", "Doc3", 45),
    ("Doc2", "Doc4", 62),
    ("Doc3", "Doc5", 30),
    ("Doc4", "Doc5", 55),
]

# Crear el grafo
G = nx.Graph()

for doc1, doc2, peso in simulaciones:
    G.add_edge(doc1, doc2, weight=peso)

# Posiciones para el grafo
pos = nx.spring_layout(G, seed=42)

# Dibujar nodos y etiquetas
plt.figure(figsize=(10, 6))
nx.draw_networkx_nodes(G, pos, node_color="skyblue", node_size=2000)
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

# Dibujar aristas
nx.draw_networkx_edges(G, pos, width=2, edge_color="gray")

# Etiquetas de peso (similitud)
etiquetas = {(u, v): f"{d['weight']}%" for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=etiquetas, font_color='red', font_size=10)

plt.title("🔍 Grafo de Similitud entre Documentos", fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```
